```python
## Explanatory Data Analysis
##Suicides_no 
ax= sns.countplot(x='sex',data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

From the above graph , we see that there are equal number of male and females in number of suicides . There is no difference in between. Lets do futher analysis to find the patterns in Age . 

```python

newdf=df.groupby(['suicides_no'], as_index=False)
df.sort_values(by='age', ascending= False)
sns.barplot(x='suicides_no',y= 'age', data= newdf.head(200), hue="sex")

```

From the above graph, we can see that number of suicides were in age group 35-54 years where female count was more than male. Overall we can see that in all age group , female count was more than male. 

```python
df.c = df.groupby(['country'], as_index=False)['suicides_no'].sum().sort_values(by='suicides_no', ascending= False).head(200)
sns.set(rc={'figure.figsize':(100,10)})
sns.barplot(data=df.c, x='country', y='suicides_no')
```

We can see that , the top  5 states in suicide rates are:

1.  United States 
2.  Japan
3. France
4. Germany
5.Ukaraine



```python
df.y = df.groupby(['year'], as_index=False)['suicides_no'].sum().sort_values(by='suicides_no', ascending= False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=df.y, x='year', y='suicides_no')
```
